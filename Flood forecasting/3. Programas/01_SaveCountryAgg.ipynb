```python
def replace_multiple_spaces(text):
  """
  Replaces multiple blank spaces with a single space in a string.

  Args:
    text: The input string.

  Returns:
    The string with multiple spaces replaced by single spaces.
  """
  return ' '.join(text.split())
```

```python
all_files_in_path_ctry_avg = os.listdir(el_path_main/f'2. Datos/Country-level spatial averages/')
file_prefix_ctry_avg = 'file_base_'
files_to_convert_ctry_avg = [entry for entry in all_files_in_path_ctry_avg if entry.startswith(file_prefix_ctry_avg)]
print(f'There where {len(files_to_convert_ctry_avg)} files found.')
```

```python
cases_not_to_include = ['MAM',
                        'JJA',
                        'SON',
                        'DJF',
                        'ANN']
map_name_number_mth = {'JAN':'01', 'FEB':'02', 'MAR':'03', 'APR':'04',
                       'MAY':'05', 'JUN':'06', 'JUL':'07', 'AUG':'08',
                       'SEP':'09', 'OCT':'10', 'NOV':'11', 'DEC':'12'}

dict_frames = {}
for x in files_to_convert_ctry_avg:
    df = pd.read_excel(el_path_main/f'2. Datos/Country-level spatial averages/{x}')
    # Get type of information that is beeing load (tmn, vap, pre, etc.)
    var_type = x[10:13]
    ctry_name = x.split('_')[-1].split('.')[0]

    df = df[0].str.strip()
    df = df.apply(replace_multiple_spaces)
    df = df.str.split(' ', expand=True)
    
    las_cols = df.iloc[0,:]
    el_cont = df.iloc[1:,:]

    df = el_cont.copy()
    df.columns = las_cols
    df = df.melt(id_vars = 'YEAR')
    df.columns = ['YEAR', 'MONTH', var_type.upper()]

    # Drop cases that are not relevant
    df = df[~df['MONTH'].isin(cases_not_to_include)]

    # Set country name
    df['COUNTRY'] = ctry_name

    # Transform month name to number
    df['MONTH'] = df['MONTH'].map(map_name_number_mth)

    # Codmes creation
    df['CODMES'] = df['YEAR']+df['MONTH']

    df.drop(['YEAR', 'MONTH'], axis=1, inplace=True)

    if var_type not in dict_frames:
        dict_frames[var_type] = {}
        dict_frames[var_type][ctry_name] = df.copy()
    else:
        dict_frames[var_type][ctry_name] = df.copy()
```

```python
df_consol = pd.DataFrame(columns=['CODMES', 'COUNTRY'])
for info_type in dict_frames.keys():
    df_consol_ctries = pd.DataFrame()
    for ctry in dict_frames[info_type].keys():
        df_consol_ctries = pd.concat([df_consol_ctries, dict_frames[info_type][ctry]])
        
    df_consol = pd.merge(df_consol, df_consol_ctries,
                         on = ['CODMES', 'COUNTRY'], 
                         how='outer')
```

```python
lista_inicial_cols_df = df_consol.columns
```

```python
for col in [x for x in df_consol.columns if x not in ['CODMES', 'COUNTRY']]:
    df_consol[f'es_miss_{col}'] = df_consol[col]=='-999.0'
df_consol[[x for x in df_consol.columns if x.startswith('es_miss_') ]].mean()
# No hay missings
```

```python
df_consol = df_consol[lista_inicial_cols_df]
```

```python
for col in [x for x in df_consol.columns if x not in ['CODMES', 'COUNTRY']]:
    df_consol[col] = pd.to_numeric(df_consol[col], errors='coerce')
```
